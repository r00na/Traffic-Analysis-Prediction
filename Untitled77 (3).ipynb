Converting time and extracting hour and minute from the time column

```python
df['Time'] = pd.to_datetime(df['Time'], format='%I:%M:%S %p', errors='coerce').dt.time
df['Hour'] = df['Time'].apply(lambda x: x.hour if pd.notna(x) else None)
df['Minute'] = df['Time'].apply(lambda x: x.minute if pd.notna(x) else None)

df[['Time', 'Hour', 'Minute']].head()
```

Parsing and extracting day, month, year, and weekday from a date column

```python
year = 2025
month = 1

df['Date'] = pd.to_datetime(df['Date'].astype(str) + f'/{month}/{year}', format='%d/%m/%Y')

df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df['DayOfWeek'] = df['Date'].dt.weekday

df[['Date', 'Day', 'Month', 'Year', 'DayOfWeek']].head()
```

# Converting columns to numeric, replacing negative values with 0, and filling missing values with 0

```python
df['CarCount'] = pd.to_numeric(df['CarCount'], errors='coerce')
df['BikeCount'] = pd.to_numeric(df['BikeCount'], errors='coerce')
df['BusCount'] = pd.to_numeric(df['BusCount'], errors='coerce')
df['TruckCount'] = pd.to_numeric(df['TruckCount'], errors='coerce')

df[['CarCount', 'BikeCount', 'BusCount', 'TruckCount']] = df[['CarCount', 'BikeCount', 'BusCount', 'TruckCount']].apply(
    lambda x: x.where(x >= 0, 0))
df.fillna(0, inplace=True)

df[['CarCount', 'BikeCount', 'BusCount', 'TruckCount']].head()
```

Calculating the total count and verifying if it matches the provided 'Total' column

```python
df['TotalCheck'] = df['CarCount'] + df['BikeCount'] + df['BusCount'] + df['TruckCount']
df['TotalCorrect'] = df['Total'] == df['TotalCheck']

df[['CarCount', 'BikeCount', 'BusCount', 'TruckCount', 'Total', 'TotalCheck', 'TotalCorrect']].head()
```

Encoding the 'Traffic Situation' column using LabelEncoder

```python
le = LabelEncoder()

df['TrafficSituation'] = le.fit_transform(df['Traffic Situation'])
df[['Traffic Situation']].head()
```

Checking Missing Values

```python
missing_values = df.isnull().sum()
print("Missing Values in each column:\n", missing_values)
```

detect outliersRemoving outliers from numeric columns using the IQR method and plotting box plots to visualize the cleaned data

```python
numeric_columns = ['CarCount', 'BikeCount', 'BusCount', 'TruckCount', 'Total']

Q1 = df[numeric_columns].quantile(0.25)
Q3 = df[numeric_columns].quantile(0.75)
IQR = Q3 - Q1

df_no_outliers = df[~((df[numeric_columns] < (Q1 - 1.5 * IQR)) | (df[numeric_columns] > (Q3 + 1.5 * IQR))).any(axis=1)]

df_no_outliers.head()
columns = ['CarCount', 'BikeCount', 'BusCount', 'TruckCount', 'Total']
plt.figure(figsize=(10, 6))

import matplotlib.pyplot as plt

columns = ['CarCount', 'BikeCount', 'BusCount', 'TruckCount', 'Total']
plt.figure(figsize=(10, 6))

for i, col in enumerate(columns, 1):
    plt.subplot(2, 3, i)
    df_no_outliers[col].plot(kind='box', title=col)
    plt.tight_layout()

plt.show()
```

# Correlations between different vehicle types

```python
vehicle_cols = ['CarCount', 'BusCount', 'TruckCount', 'BikeCount_sqrt', 'Total']

corr_matrix = df[vehicle_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="Blues", fmt=".2f", linewidths=0.5)

plt.title("Correlation Between Different Vehicle Types")
plt.show()
```

1. Strong positive correlation between CarCount and Total, indicating that cars are the dominant vehicle type affecting total traffic.  
2. TruckCount and BusCount may have a weaker correlation, suggesting they vary independently depending on time or location.  
3. BikeCount may show lower correlation with other vehicle types, as bike movement patterns differ from motor vehicles.

# The variance in vehicle counts compare across vehicle types

```python
vehicle_types = ['CarCount', 'BusCount', 'TruckCount', 'BikeCount_sqrt']
variances = [np.var(df[col]) for col in vehicle_types]

plt.figure(figsize=(8, 5))
sns.barplot(x=vehicle_types, y=variances, palette="Blues")

plt.xlabel("Vehicle Type")
plt.ylabel("Variance")
plt.title("Variance in Vehicle Counts Across Vehicle Types")
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

1. CarCount is likely to have the highest variance due to the dominance of private vehicles.
2. BusCount and TruckCount may have lower variance, as they operate on fixed schedules and routes.
3. BikeCount may show less variance, as bicycle traffic depends more on weather and infrastructure.
